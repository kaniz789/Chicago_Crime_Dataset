# **Analysis:**

#**1.	Crime Trends and Seasonality**

---



# **Question:** Overall Crime Rate Trend: What are the overall trends in crime rates over time?

---



# Group by Year and Count: Group the data by year and calculate the total number of crimes for each year.

```python
dataset_df['Date'] = pd.to_datetime(dataset_df['Date'])
```

```python
yearly_crime_counts = dataset_df.groupby(dataset_df['Date'].dt.year)['ID'].count().reset_index()
yearly_crime_counts.columns = ['Year', 'Crime_Count']
```

```python
import matplotlib.pyplot as plt

plt.plot(yearly_crime_counts['Year'], yearly_crime_counts['Crime_Count'])
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.title('Overall Crime Rate Trend')
plt.show()
```

# **Crime Seasonality**

---
# Group by Month and Count: Group the data by month and calculate the total number of crimes for each month.

```python
monthly_crime_counts = dataset_df.groupby(dataset_df['Date'].dt.month)['ID'].count().reset_index()
monthly_crime_counts.columns = ['Month', 'Crime_Count']
```

```python
plt.plot(monthly_crime_counts['Month'], monthly_crime_counts['Crime_Count'])
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.title('Crime Seasonality')
plt.show()
```

# **Analyze Daily Crime Counts:**
---
# **Question:** Are there seasonal patterns, such as higher incidents in certain months or days of the week?
---
# To analyze daily patterns, we can group the data by the day of the week and count the crimes.

```python
daily_crime_counts = dataset_df.groupby(dataset_df['Date'].dt.dayofweek)['ID'].count().reset_index()
daily_crime_counts.columns = ['DayOfWeek', 'Crime_Count']

# Map day of week numbers to day names for better readability
day_names = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_crime_counts['DayOfWeek'] = daily_crime_counts['DayOfWeek'].apply(lambda x: day_names[x])

# Plot the daily crime counts
plt.plot(daily_crime_counts['DayOfWeek'], daily_crime_counts['Crime_Count'])
plt.xlabel('Day of Week')
plt.ylabel('Number of Crimes')
plt.title('Crime Seasonality by Day of Week')
plt.show()
```

# **Question:** 	Are some crime types more likely to result in arrests?

```python
# Create a new column 'Arrest_Made' indicating if an arrest was made
dataset_df['Arrest_Made'] = dataset_df['Arrest'].map({True: 1, False: 0})

# Group by crime type and calculate the arrest rate
arrest_rates = dataset_df.groupby('Primary Type')['Arrest_Made'].mean().reset_index()
arrest_rates.columns = ['Primary Type', 'Arrest_Rate']

# Sort by arrest rate in descending order
arrest_rates = arrest_rates.sort_values('Arrest_Rate', ascending=False)

# Display the arrest rates for each crime type
print(arrest_rates)
```

```python
# Create a bar plot to visualize the arrest rates
plt.figure(figsize=(12, 6))
sns.barplot(x='Arrest_Rate', y='Primary Type', data=arrest_rates.head(10), palette='viridis')
plt.title('Arrest Rates for Top 10 Crime Types')
plt.xlabel('Arrest Rate')
plt.ylabel('Crime Type')
plt.show()
```

```python
# Group by 'Ward' and count crime incidents
ward_crime_counts = dataset_df.groupby('Ward')['ID'].count().reset_index()
ward_crime_counts.columns = ['Ward', 'Crime_Count']

# Assuming you have population data for each ward in a separate DataFrame
# called 'ward_population', you need to create or load it here.
# For example, if you have the data in a CSV file:
# ward_population = pd.read_csv('ward_population.csv')

# Placeholder for ward population data (replace with your actual data)
# Creating a sample ward_population DataFrame for demonstration:
ward_population = pd.DataFrame({'Ward': ward_crime_counts['Ward'],
                                   'Population': np.random.randint(10000, 50000, size=len(ward_crime_counts))})

# Now you can proceed with the merge:
ward_crime_rates = pd.merge(ward_crime_counts, ward_population, on='Ward')

# Calculate crime rate per capita (crimes per 1000 people)
ward_crime_rates['Crime_Rate'] = (ward_crime_rates['Crime_Count'] / ward_crime_rates['Population']) * 1000
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Ward', y='Crime_Rate', data=ward_crime_rates)
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.title('Crime Rates by Ward')
plt.xlabel('Ward')
plt.ylabel('Crime Rate (per 1000 people)')
plt.show()
```
